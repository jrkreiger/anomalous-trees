# Get & prep data

```python
# Import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (confusion_matrix, classification_report, auc, 
                             roc_curve, accuracy_score)
from sklearn.svm import OneClassSVM
from sklearn.preprocessing import StandardScaler

from imblearn.combine import SMOTEENN
```

```python
# Define a function to visualize a confusion matrix
def pretty_confusion(y_true, y_pred, class_names, model_name):
    '''Display normalized confusion matrix with color scale.
    
    Note that this function works specifically for data from the 
    Forest Cover Types dataset. 
    
    Keyword arguments:
    y_true: ground-truth labels
    y_pred: predicted labels
    class_names: list of class names
    model_name: name to print in the plot title
    
    Dependencies:
    numpy aliased as np
    sklearn.metrics.confusion_matrix
    matplotlib.pyplot aliased as plt
    seaborn aliased as sns
    '''
    
    # Calculate the confusion matrix
    matrix = confusion_matrix(y_true, y_pred)
    matrix = matrix.astype('float') / matrix.sum(axis=1)[:, np.newaxis]
    
    # Build the plot
    plt.figure(figsize=(10,6))
    sns.set(font_scale=1.4)
    sns.heatmap(matrix, annot=True, annot_kws={'size':10},
                cmap=plt.cm.Greens, linewidths=0.2)
    
    # Add labels to the plot
    tick_marks = np.arange(len(class_names))
    tick_marks2 = tick_marks + 0.5
    plt.xticks(tick_marks, class_names)
    plt.yticks(tick_marks2, class_names)
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.title('Confusion Matrix for {}'.format(model_name)) 
    plt.tight_layout()
```

```python
# Load original data
original = pd.read_table("~/Documents/fs-project-4/covtype.data", sep=',', header=None)

# Name the columns
colnames = ['Elevation', 'Aspect', 'Slope', 'Hor_Dist_Hyd', 'Ver_Dist_Hyd',
            'Hor_Dist_Road', 'Hillshade_9am', 'Hillshade_noon', 'Hillshade_3pm',
            'Hor_Dist_Fire', 'WA_1', 'WA_2', 'WA_3', 'WA_4', 'Soil_Type_1', 
            'Soil_Type_2', 'Soil_Type_3', 'Soil_Type_4', 'Soil_Type_5', 
            'Soil_Type_6', 'Soil_Type_7', 'Soil_Type_8', 'Soil_Type_9', 
            'Soil_Type_10', 'Soil_Type_11', 'Soil_Type_12', 'Soil_Type_13', 
            'Soil_Type_14', 'Soil_Type_15', 'Soil_Type_16', 'Soil_Type_17', 
            'Soil_Type_18', 'Soil_Type_19', 'Soil_Type_20', 'Soil_Type_21', 
            'Soil_Type_22', 'Soil_Type_23', 'Soil_Type_24', 'Soil_Type_25', 
            'Soil_Type_26', 'Soil_Type_27', 'Soil_Type_28', 'Soil_Type_29', 
            'Soil_Type_30', 'Soil_Type_31', 'Soil_Type_32', 'Soil_Type_33', 
            'Soil_Type_34', 'Soil_Type_35', 'Soil_Type_36', 'Soil_Type_37', 
            'Soil_Type_38', 'Soil_Type_39', 'Soil_Type_40', 'Cover_Type',]
original.columns = colnames

# View class counts
original['Cover_Type'].value_counts()
```

```python
# Create anomalous dataset
data = original[(original['Cover_Type'] == 2) | (original['Cover_Type'] == 5)]
data['Cover_Type'].value_counts()
```

```python
# Split into features and target
features = data.drop('Cover_Type', axis=1)
target = data['Cover_Type']
```

```python
# Get size of validation & test sets (10% of total)
val_len = int(round(0.1*len(data)))
```

```python
# Create training, validation, and test sets
X_train, X_test, y_train, y_test = train_test_split(features, target, 
                                                    test_size=val_len,
                                                    stratify=target,
                                                    random_state=1)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train,
                                                  test_size=val_len,
                                                  stratify=y_train,
                                                  random_state=1)
```

# Try to detect both classes

```python
# Fit baseline Random Forest
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)
```

```python
# Evaluate baseline Random Forest
pred = rf.predict(X_val)
pretty_confusion(y_val, pred, [2, 5], 'Baseline Random Forest')
```

```python
# View classification report for baseline Random Forest
print(classification_report(y_val, pred))
```

```python
# Fit Random Forest with classes weighted 1:5
rf = RandomForestClassifier(n_estimators=100, class_weight={2:1, 5:5})
rf.fit(X_train, y_train)

# Evaluate Random Forest
pred = rf.predict(X_val)
pretty_confusion(y_val, pred, [2, 5], 
                 'Random Forest with classes weighted 1:5')
```

```python
# View classification report for Random Forest
print(classification_report(y_val, pred))
```

```python
# Fit Random Forest with classes weighted 1:10
rf = RandomForestClassifier(n_estimators=100, class_weight={2:1, 5:10})
rf.fit(X_train, y_train)

# Evaluate Random Forest
pred = rf.predict(X_val)
pretty_confusion(y_val, pred, [2, 5], 
                 'Random Forest with classes weighted 1:10')
```

```python
# View classification report for Random Forest
print(classification_report(y_val, pred))
```

```python
# Fit Random Forest with balanced classes
smoteenn = SMOTEENN() 
X_resampled, y_resampled = smoteenn.fit_sample(X_train, y_train)
```

```python
# Fit a Random Forest
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_resampled, y_resampled)

# Evaluate Random Forest
pred = rf.predict(X_val)
pretty_confusion(y_val, pred, [2, 5], 
                 'Random Forest with classes balanced using SMOTEENN')
```

```python
# View classification report for Random Forest
print(classification_report(y_val, pred))
```

```python
# Define a function to visualize feature importances
def plot_feature_importances(model, x):
    '''Display barplot representing a model's feature importances.
    
    Note that in a Jupyter Notebook, running this function will display
    the plot automatically.
    
    Keyword arguments:
    model: a trained classifier with .feature_importances_ attribute
    x: the training dataset or other object of identical dimensions
    
    Dependencies:
    matplotlib.pyplot aliased as plt
    '''
    n_features = x.shape[1]
    plt.figure(figsize=(8,10))
    plt.barh(range(n_features), model.feature_importances_, align='center') 
    plt.yticks(np.arange(n_features), range(1, n_features+1)) 
    plt.xlabel("Feature importance")
    plt.ylabel("Feature (Component)")
    plt.title('Feature Importances')
```

```python
# Visualize feature importances
plot_feature_importances(rf, X_train)
```
